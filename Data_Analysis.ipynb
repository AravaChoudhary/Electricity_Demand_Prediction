```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("data.csv")
```

```python
dataframe
```

```python
dataframe.describe()
```

```python
#Bar chart for hour_of_day vs electricity load
plt.figure(figsize=(10, 6))
sns.barplot(x='hour_of_day', y='electricity_load_MW', data=dataframe, estimator=sum, ci=None)
plt.title('Electricity Load by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Total Electricity Load (MW)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line plot for hour_of_day vs electricity load
plt.figure(figsize=(10, 6))
sns.lineplot(x='hour_of_day', y='electricity_load_MW', data=dataframe, estimator='sum', ci=None)
plt.title('Electricity Load by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Total Electricity Load (MW)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Line plot for is_holiday vs electricity load
plt.figure(figsize=(6, 6))
sns.lineplot(x='is_holiday', y='electricity_load_MW', data=dataframe, estimator='sum', ci=None)
plt.title('Electricity Load on Holidays vs Non-Holidays')
plt.xlabel('Holiday (1 = Yes, 0 = No)')
plt.ylabel('Total Electricity Load (MW)')
plt.show()
```

```python
# Boxplot for electricity load across different seasons
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='electricity_load_MW', data=dataframe)
plt.title('Electricity Load Distribution Across Seasons')
plt.xlabel('Season')
plt.ylabel('Electricity Load (MW)')
plt.show()
```

```python
# Boxplot for electricity load during holidays vs non-holidays
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_holiday', y='electricity_load_MW', data=dataframe)
plt.title('Electricity Load During Holidays vs Non-Holidays')
plt.xlabel('Holiday (1 = Yes, 0 = No)')
plt.ylabel('Electricity Load (MW)')
plt.show()
```

```python
#Pie chart for hour_of_day: Proportion of electricity load by hour of the day
hourly_data = dataframe.groupby('hour_of_day')['electricity_load_MW'].sum()

plt.figure(figsize=(8, 8))
plt.pie(hourly_data, labels=hourly_data.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Electricity Load by Hour of Day')
plt.show()
```

```python
# Pie chart for is_holiday: Proportion of electricity load during holidays vs non-holidays
holiday_data = dataframe.groupby('is_holiday')['electricity_load_MW'].sum()

plt.figure(figsize=(6, 6))
plt.pie(holiday_data, labels=['Non-Holiday', 'Holiday'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Proportion of Electricity Load on Holidays vs Non-Holidays')
plt.show()
```
